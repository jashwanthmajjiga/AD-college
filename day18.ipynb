```python
from scipy import stats
import pandas as pd

# Example DataFrame
data = {'A': [10, 12, 15, 100, 18, 20, 22, 105]}  # Contains outliers
df = pd.DataFrame(data)

# Compute Z-scores
z_score = stats.zscore(df)

# Identify outliers (absolute Z-score >= 3)
outliers = (abs(z_score) >= 3).any(axis=1)

# Select rows that are outliers
outliers_rows = df[outliers]

print(outliers_rows)
```
